```python
def opimage(image_paths,image_classes):
    imgs = []
    time=1
    allt=len(image_paths)
    print("Calculating all the descriptors and keypoints...")
    for image_path in image_paths:
        img  = cv.imread(image_path)
        img = cv.cvtColor(img,cv.COLOR_BGR2GRAY) # Converts from one color space to another (grayscale)
        img =cv2.resize(img,(32,32),interpolation=cv2.INTER_CUBIC)
        #kp, des = sift.detectAndCompute(img,None)
        imgs.append((img))
        print(str(time)+"/"+str(allt), end="\r")
        time=time+1
    print("\nDone!")
    return(imgs,image_paths,image_classes)   
```

```python
def sift(imgs,image_paths,image_classes):
    sift= cv.xfeatures2d.SIFT_create()
    des_list = []
    time=1
    allt=len(imgs)
    for im in imgs:
        kp, des = sift.detectAndCompute(im, None)
        if str(des) != 'None':
            des_list.append((image_paths[time-1],des))
            print(str(time)+"/"+str(allt), end="\r")
            time=time+1
        else:
            print(image_paths[time-1])
            del image_classes[time-1]
            del image_paths[time-1]
    print("\nDone!")
    
    
    # Stack all the descriptors vertically in a numpy array
    print("Stacking all the descriptors in a numpy array...")
    time=1
    allt=len(des_list)
    descriptors = des_list[0][1]
    for image_path, descriptor in des_list[1:]:
        print(str(time)+"/"+str(allt), end="\r")
        time=time+1
        descriptors = np.vstack((descriptors, descriptor))
    print("\nDone!")
    k = 128
    voc, variance = kmeans(descriptors, k, 1)
    print("Done doing K-means")
    im_features = np.zeros((len(image_paths), k), "float32")
    for i in range(len(image_paths)):
        print("Calculating distance for image "+str(i)+"...", end="\r")
        words, distance = vq(des_list[i][1],voc)
        for w in words:
            im_features[i][w] += 1
    # Perform Tf-Idf vectorization
    nbr_occurences = np.sum( (im_features > 0) * 1, axis = 0)
    idf = np.array(np.log((1.0*len(image_paths) + 1) / (1.0*nbr_occurences + 1)), 'float32')
    scaled_im_features = im_features * idf
    stdSlr = StandardScaler().fit(scaled_im_features)
    scaled_im_features = stdSlr.transform(scaled_im_features)
    return(scaled_im_features,image_paths,image_classes,stdSlr,k,voc)
def lbp1(imgs,image_paths,image_classes):
    radius = 1  # LBP算法中範圍半徑的取值
    n_points = 8 * radius # 領域像素點數
    des_list = []
    time=1
    allt=len(imgs)
    for im in imgs:
        lbp = local_binary_pattern(im, n_points, radius)
        max_bins = int(lbp.max() + 1)
        des,_ = np.histogram(lbp,bins=max_bins, range=(0, max_bins))
        if str(des) != 'None':
            des_list.append((des))
            print(str(time)+"/"+str(allt), end="\r")
            time=time+1
        else:
            print(image_paths[time-1])
            del image_classes[time-1]
            del image_paths[time-1]
    print("\nDone!")
   # return(scaled_im_features,image_paths,image_classes,stdSlr,k,voc)
    return(des_list,image_paths,image_classes)
```

```python
def svcf(scaled_im_features,image_classes):
    clf = LinearSVC()
    clf.fit(scaled_im_features, np.array(image_classes))
    return(clf)
```
